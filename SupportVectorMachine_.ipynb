# Support Vector Machine (SVM)

Support-Vector Machines (SVMs, also support-vector network) are supervised learning models with associated learning algorithms that analyze data for classification and regression analysis. Developed at AT&T Bell Laboratories by Vladimir Vapnik with his colleagues.

A support-vector machine constructs a hyperplane or set of hyperplanes in a high- or infinite-dimensional space, which can be used for classification, regression, or other tasks like outliers detection. Intuitively, a good separation is achieved by the hyperplane that has the largest distance to the nearest training-data point of any class (so-called functional margin), since in general the larger the margin, the lower the generalization error of the classifier.

**we'll be using an in-built dataset of sklearn library known by the name Wisconsin Breast Cancer data**

* import dependencies(libraries)
* Loading the data 

***The DataFrame will have only the independent variables dependent variables are assigned by data.target function***

```python
import pandas as pd 
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

df = pd.DataFrame(cancer.data, columns=[cancer.feature_names])
df
```

```python
# Target variable 

cancer.target[1:50]
```

```python
# Statistical Summary of the Data
df.describe()
```

Since the data is in-built it won't have null values or outliers ,to some extent the data will be cleaned and preprocessed 

* We will see if theres any null values 

```python
df.isna().sum()
```

Next we will split the data for training and testing accordingy using sklaern functions 

```python
X = df # The DataFrame having all the independent features(variables)
Y = cancer.target # Target variable  

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y ,random_state = 23, stratify = Y)
```

Till now we saw the data and its features now we create the SVm model object 

* import model
* fit the model
* get predictions 
* Evaluate model performnace 

```python
import warnings 
warnings.filterwarnings('ignore') # To hide all possible warnings 

from sklearn.svm import SVC # "Support vector classifier"
model = SVC(kernel = 'linear') # Binary classification roblem
model.fit(x_train, y_train)
y_pred_svm = model.predict(x_test)
y_pred_svm
```

```python
# Model evaluation 
from sklearn.metrics import accuracy_score , classification_report 
accuracy_svm = accuracy_score(y_pred_svm,y_test)
accuracy_svm
```

```python
print(classification_report(y_pred_svm,y_test))
```
